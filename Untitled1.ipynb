```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ==========================================
# 1. DATA ACQUISITION & PREPARATION
# ==========================================
def fetch_historical_data(start_date='2016-01-01', end_date='2026-06-24'):
    print("Fetching historical data...")
    # SI=F: COMEX Silver Futures (Global Benchmark)
    # DX-Y.NYB: US Dollar Index (Global Macro Cue)
    # INR=X: USD/INR Exchange Rate (Indian Market Cue)
    
    tickers = ['SI=F', 'DX-Y.NYB', 'INR=X']
    
    # Using 'Close' as modern yfinance adjusts prices directly into 'Close'
    data = yf.download(tickers, start=start_date, end=end_date)['Close']
    
    # Modern pandas forward fill syntax
    data = data.ffill()
    data.dropna(inplace=True)
    
    # Calculate synthetic "Indian Silver" price proxy (Global Silver * USD/INR)
    data['MCX_Proxy'] = data['SI=F'] * data['INR=X']
    
    return data

# ==========================================
# 2. SIGNAL GENERATION (TECH + MACRO)
# ==========================================
def generate_signals(df):
    print("Generating trading signals with Macro Filters...")
    
    # Moving Averages for Trend Identification
    df['SMA_20'] = df['MCX_Proxy'].rolling(window=20).mean()
    df['SMA_50'] = df['MCX_Proxy'].rolling(window=50).mean()
    
    # MACRO FILTER: Dollar Index Trend (Inverse to Silver typically)
    # If Dollar is above its 50-day SMA, macro sentiment is "Bearish" for Silver.
    df['DXY_SMA_50'] = df['DX-Y.NYB'].rolling(window=50).mean()
    df['Macro_Bearish'] = df['DX-Y.NYB'] > df['DXY_SMA_50']
    
    # Generate Base Signals
    df['Signal'] = 0 
    
    # Rule 1: Bullish Trend (Fast MA > Slow MA) AND Dollar is Weak (Macro filter)
    bull_condition = (df['SMA_20'] > df['SMA_50']) & (~df['Macro_Bearish'])
    
    # Rule 2: Bearish Trend (Fast MA < Slow MA) AND Dollar is Strong (Macro filter)
    bear_condition = (df['SMA_20'] < df['SMA_50']) & (df['Macro_Bearish'])
    
    df.loc[bull_condition, 'Signal'] = 1  # Buy / Bull Call Spread territory
    df.loc[bear_condition, 'Signal'] = -1 # Sell / Bear Put Spread territory
    
    # Shift signals by 1 day to avoid look-ahead bias (we trade the day AFTER signal)
    df['Signal'] = df['Signal'].shift(1)
    
    return df.dropna()

# ==========================================
# 3. BACKTEST ENGINE & METRICS
# ==========================================
def run_backtest(df):
    print("Running Backtest...")
    
    # Calculate daily returns of the asset
    df['Daily_Return'] = df['MCX_Proxy'].pct_change()
    
    # Calculate Strategy Returns
    df['Strategy_Return'] = df['Signal'] * df['Daily_Return']
    
    # Cumulative Returns
    df['Cumulative_Market'] = (1 + df['Daily_Return'].fillna(0)).cumprod()
    df['Cumulative_Strategy'] = (1 + df['Strategy_Return'].fillna(0)).cumprod()
    
    # Probability and Edge Metrics
    winning_days = len(df[df['Strategy_Return'] > 0])
    losing_days = len(df[df['Strategy_Return'] < 0])
    total_trades = winning_days + losing_days
    
    win_rate = winning_days / total_trades if total_trades > 0 else 0
    
    # Annualized Volatility and Sharpe Ratio
    annualized_return = df['Strategy_Return'].mean() * 252
    annualized_vol = df['Strategy_Return'].std() * np.sqrt(252)
    sharpe_ratio = annualized_return / annualized_vol if annualized_vol > 0 else 0
    
    print("\n" + "="*30)
    print("BACKTEST RESULTS (10 YEARS)")
    print("="*30)
    print(f"Total Trading Days Filtered: {total_trades}")
    print(f"Win Rate (Probability of Success): {win_rate:.2%}")
    print(f"Annualized Return: {annualized_return:.2%}")
    print(f"Sharpe Ratio (Risk-Adjusted Edge): {sharpe_ratio:.2f}")
    print(f"Cumulative Market Return: {df['Cumulative_Market'].iloc[-1]:.2f}x")
    print(f"Cumulative Strategy Return: {df['Cumulative_Strategy'].iloc[-1]:.2f}x")
    print("="*30)
    
    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(df.index, df['Cumulative_Market'], label='Buy & Hold (MCX Proxy)', color='gray', alpha=0.7)
    plt.plot(df.index, df['Cumulative_Strategy'], label='Macro-Filtered Strategy', color='blue', linewidth=2)
    plt.title('10-Year Backtest: Macro-Filtered Silver Strategy')
    plt.xlabel('Date')
    plt.ylabel('Cumulative Return Multiplier')
    plt.legend()
    plt.grid(True)
    plt.show()

# ==========================================
# 4. EXECUTION
# ==========================================
if __name__ == "__main__":
    raw_data = fetch_historical_data('2016-01-01', '2026-06-24')
    signal_data = generate_signals(raw_data)
    run_backtest(signal_data)
```
